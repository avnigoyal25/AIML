Q8 Plot a graph showing the distribution of income in different age groups. For example, how many people between the ages of 21-30 earn more than 50k and how many earn less than 50k. You have to plot one graph for all the age groups.

```python
def age_group(x):
    x = int(x)
    x = abs(x)
    if( 18 < x < 31 ):
        return "19-30"
    if( 30 < x < 41 ):
        return "31-40"
    if( 40 < x < 51 ):
        return "41-50"
    if( 50 < x < 61 ):
        return "51-60"
    if( 60 < x < 71 ):
        return "61-70"
    else:
        return "Greater than 70"

df['age_group'] = df['Age'].apply(age_group)
```

```python
# Your code goes here

plt.figure(figsize=(12,6))
order_list = ['19-30', '31-40', '41-50', '51-60', '61-70', 'Greater than 70']
sns.countplot(x=df['age_group'], hue = df['income'], palette='Greens_r', order = order_list)
plt.title('Income of Individuals of Different Age Groups', fontsize=18, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
```

Q9 Plot a similar graph for hours per week and income.

```python
# Your code goes here
def hours(x):
    x = int(x)
    x = abs(x)
    if( 1 < x < 21 ):
        return "1-20"
    if( 20 < x < 41 ):
        return "21-40"
    if( 40 < x < 61 ):
        return "41-60"
    if( 60 < x < 81 ):
        return "61-80"
    else:
        return "Greater than 80"

df['hours'] = df['hours per week'].apply(hours)
```

```python
plt.figure(figsize=(12,6))
order_list = ['1-20', '21-40', '41-60', '61-80', 'Greater than 80']
sns.countplot(x=df['hours'], hue = df['income'], palette='Greens_r', order = order_list)
plt.title('Income of Individuals of Different Age Groups', fontsize=18, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
```
